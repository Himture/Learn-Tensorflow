```python
from transformers import ViTFeatureExtractor, ViTForImageClassification
from PIL import Image as img
from IPython.display import Image, display
```

```python
FILE_NAME = 'https://media.babolat.com/image/upload/f_auto,q_auto,c_crop,w_810,h_363/Website_content/Homepage/02092020-Launch/PA_Rafa_810_x_364_disclaimer.png'
```

```python
!wget 'https://media.babolat.com/image/upload/f_auto,q_auto,c_crop,w_810,h_363/Website_content/Homepage/02092020-Launch/PA_Rafa_810_x_364_disclaimer.png'
```

```python
display(Image(FILE_NAME, width = 700, height = 400))
```

```python
image_array = img.open('/content/PA_Rafa_810_x_364_disclaimer.png')
```

```python
feature_extractor = ViTFeatureExtractor.from_pretrained('google/vit-base-patch16-224')
model = ViTForImageClassification.from_pretrained('google/vit-base-patch16-224')
```

```python
inputs = feature_extractor(images = image_array,return_tensors="pt")
outputs = model(**inputs)
logits = outputs.logits
```

```python
predicted_class_idx = logits.argmax(-1).item()
```

```python
print(predicted_class_idx)
```

```python
print("Predicted class:", model.config.id2label[predicted_class_idx])
```

```python
from transformers import ViTFeatureExtractor, ViTForImageClassification
from PIL import Image
import requests
```

```python
#from IPython.display import Image
display(image_array)
```

```python
feature_extractor = ViTFeatureExtractor.from_pretrained('google/vit-base-patch16-224')
model = ViTForImageClassification.from_pretrained('google/vit-base-patch16-224')
```

```python
inputs = feature_extractor(images=image_array, return_tensors="pt")
```

```python
outputs = model(**inputs)
```

```python
logits = outputs.logits
```

```python

predicted_class_idx = logits.argmax(-1).item()
print("Predicted class:", model.config.id2label[predicted_class_idx])
```
